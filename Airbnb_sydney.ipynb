### **2.3 - Tratando Outliers**


**O que são outliers?**

>Os outliers são dados que se diferenciam drasticamente de todos os outros. Em outras palavras, um outlier é um valor que foge da normalidade e que pode (e provavelmente irá) causar anomalias, nos resultados obtidos por meio de algoritmos e sistemas de análise.

**Razões para isso.**

* Entrada de dados incorreta ou erro durante o processamento de dados.
* Erros ocorrem durante os experimentos.
* Não é um erro, seria incomum do original.
* Distribuição extrema do que o normal.

```python
df_clean.describe()
```

**Encontrando Outliers:**

Usamos o histograma para termos uma visão geral dos outliers. Para o tratamentos de Outliers podemos excluí-los ou não.

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax0.hist('price', data=df_clean, bins=25, density=True)
ax0.set_title("Histograma da preço")
ax0.set_ylabel("Quantidade de ocorrência(%)")
ax0.set_xlabel("Preço(Dólar australiano)")


ax1.hist('minimum_nights', data=df_clean, bins=25, density=True)
ax1.set_title("Histograma de minímos de noites")
ax1.set_ylabel("Quantidade de ocorrências(%)")
ax1.set_xlabel("Número de noites")

plt.tight_layout()
```

```python
#plotar boxplot
fig, ax = plt.subplots(figsize=(20,5))
sns.set_theme(style="whitegrid")
ax.set_title("boxplot de preços")
ax = sns.boxplot(x=df_clean.price)
```

Ao analisar o boxplot, podemos confirmar que existem outliers que atraplham. Usaremos a variação interquartil.

**Ultilazeremos a variação interquartil (IQR) que funciona assim:**

Pegamos o terceiro quartil q3 e subtraímos pelo o primeiro quartil q1 
* iqr = q3 - q1.

Agora identificaremos o limite inferior e o limite superior.

* limite inferior = q1 - 1.5 * iqr
* limite superior = q3 + 1.5 * iqr

```python
#Definir os quartis.
q1_price = df_clean.price.quantile(0.25)
q3_price = df_clean.price.quantile(0.75)

#definir o IQR.
iqr_price = q3_price - q1_price

#cálculo.
low_price = q1_price - 1.5 * iqr_price
up_price = q3_price + 1.5 * iqr_price

print('limite inferior = ', low_price)
print('limite superior = ', up_price)
```

Como se trata de preço, não existem outliers inferiores com valores negativos nesse caso.

```python
#quantos valores estão acima do limite superior ? 
print(f"Entradas acima do limite superior: {len(df_clean[df_clean.price > up_price])} ")
print(f"esses dados representam: {len(df_clean[df_clean.price > up_price])  / df_clean.shape[0] * 100 :.2f}%")
```

**Agora faremos isso com os outros.**

```python
#plot um boxplot.
plt.figure (figsize = (20,5))
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=df_clean.minimum_nights)
```

```python
#Definir os quartis.
q1_min_nights= df_clean.minimum_nights.quantile(0.25)
q3_min_nights= df_clean.minimum_nights.quantile(0.75)

#definir o IQR.
iqr_min_nights= q3_min_nights- q1_min_nights

#cálculo.
low_min_nights= q1_min_nights- 1.5 * iqr_min_nights
up_min_nights= q3_min_nights+ 1.5 * iqr_min_nights

print('limite inferior = ', low_price)
print('limite superior = ', up_price)
```

```python
#quantos valores estão acima do limite superior ? 
print(f"Entradas acima do limite superior: {len(df_clean[df_clean.minimum_nights > up_min_nights])} ")
print(f"esses dados representam: {len(df_clean[df_clean.minimum_nights > up_min_nights])  / df_clean.shape[0] * 100 :.2f}%")
```

### **3.4 - Quais os bairros mais caros?**

Para ser uma análise mais precisa peguei apenas os bairros que tinham mais de 100 anúncios, removendo os outros para que não causem uma falsa percepção em que poucos anúncios caros aumentem a média.

```python
#definir uma novo dataframe.
df_clean3 = df_clean2[['host_neighbourhood', 'price', 'review_scores_rating']]

#adicionar uma nova coluna de frequência de vezes que um bairro aparece.
df_clean3['freq'] = df_clean3.groupby('host_neighbourhood')['host_neighbourhood'].transform('count')

#apagar os bairros que tem menos de 100 anúncios.
df_clean3.drop(df_clean3[df_clean3.freq < 100].index, axis=0, inplace=True)
```

```python
#agrupar por bairro e calcular a média
neighbourhood_mean = df_clean3.groupby(['host_neighbourhood']).mean()

#posicionar em ordem decrescente.
neighbourhood_mean = neighbourhood_mean.sort_values(by=['price'], ascending=False)
```
